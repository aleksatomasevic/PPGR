```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
import cv2
import numpy as np
import numpy.linalg as la

def projPres0(pts):
  l1 = la.det(np.array([pts[3], pts[1], pts[2]]).T);
  l2 = la.det(np.array([pts[0], pts[3], pts[2]]).T);
  l3 = la.det(np.array([pts[0], pts[1], pts[3]]).T);
  return np.array([l1*pts[0], l2*pts[1], l3*pts[2]]).T

def projPres(pts1, pts2):
  P = projPres0(pts1)
  Q = projPres0(pts2)
  return np.dot(Q, la.inv(P))
```

```python
image = mpimg.imread("IMG_0223.jpeg")
image = np.rot90(image, k=-1) 
plt.imshow(image)
plt.title("Originalna slika")
plt.show()
```

```python
tacke = np.array([
    [400, 1392, 1],
    [1648, 1217, 1],
    [2841, 2428, 1],
    [867, 3048, 1]
], dtype=float)
```

```python
#Dimenzije kao A4 papir
w = 600   # irina
h = 900   # visina
```

```python
slike_tacaka = np.array([
    [0,     0,    1],
    [w - 1, 0,    1],
    [w - 1, h - 1, 1],
    [0,     h - 1, 1]
], dtype=float)
```

```python
P = projPres(tacke, slike_tacaka)
print(P)
```

```python
#Ispravljanje distorzije
img_new = cv2.warpPerspective(image, P, (w, h))

plt.imshow(img_new)
plt.axis('off')
plt.title("Slika sa otklonjenom projektivnom distorzijom")
plt.show()
```
