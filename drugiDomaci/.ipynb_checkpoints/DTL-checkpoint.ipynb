```python
import numpy as np
import numpy.linalg as la
```

```python
O3 = np.array([0,0,0])

def dveJed(Ah, A1h):
    jed1 = np.concatenate((O3, -A1h[2] * Ah, A1h[1] * Ah))
    jed2 = np.concatenate((A1h[2] * Ah, O3, -A1h[0] * Ah))
    return np.array([jed1, jed2])
```

```python
def DLT(pts, pts1):
    A = []
    for i in range(len(pts)):
        dve = dveJed(pts[i], pts1[i])
        A.append(dve[0])
        A.append(dve[1])
    A = np.array(A) #mora opet u numpy zbog la.svd

    U, S, Vh = la.svd(A) #klasik SVD
    matP = Vh[8].reshape(3,3) 

    #normiranje da u donjem desnom uglu bude broj 1
    return matP / matP[2, 2]
```

```python
n = 1

# polazne tacke (homogene)
M1 = np.array([1.,  2.,  3.])
M2 = np.array([3.,  2.,  1.])
M3 = np.array([1., -2.,  1.])
M4 = np.array([1.,  1.,  1.])
M5 = np.array([0., -2.,  3.])
M6 = np.array([1.,  0., -1.])

# ciljne
M1p = np.array([21.3 + n,      -3.2,        7.7 + 0.4*n])
M2p = np.array([11.1 + 3*n,     1.6,       15.1 + 0.4*n])
M3p = np.array([-0.8 + n,      -6.5,        5.7 - 0.4*n])
M4p = np.array([ 8.1 + n,      -0.4,        5.8 + 0.2*n])
M5p = np.array([ 9.3,         -11.2,        3.0 - 0.4*n])
M6p = np.array([-5.1 + n,       2.4,        3.7        ])

pts  = [M1, M2, M3, M4, M5, M6]
pts1 = [M1p, M2p, M3p, M4p, M5p, M6p]

P = DLT(pts, pts1)
print(P)
print(f'p33 = {P[2][2]}')
```
