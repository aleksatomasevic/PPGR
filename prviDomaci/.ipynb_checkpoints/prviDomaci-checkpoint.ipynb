```python
import numpy as np

def u_homogene(t):
    x,y = t
    return np.array([x, y, 1.0])

def prava_kroz(a, b):
    return np.cross(a,b)

def presek_pravih(l1, l2):
    p = np.cross(l1, l2)
    if abs(p[2]) < 1e-9:
        return p
    return p / p[2]
    
```

```python
def osmoteme(lista):
    
    p1, p2, p3, p5, p6, p7, p8 = lista

    P1, P2, P3, P5, P6, P7, P8 = [u_homogene(p) for p in (p1, p2, p3, p5, p6, p7, p8)]

    #(paralelne ivice P1P2 i P5P6)
    
    BeskonacnoX = presek_pravih(
        prava_kroz(P1, P2),
        prava_kroz(P5, P6)
    )

    #(paralelne ivice P1P2 i P5P6)
    BeskonacnoY = presek_pravih(
        prava_kroz(P1, P5),
        prava_kroz(P2, P6)
    )

    # P4 je na:
    # pravoj kroz P3 u x-smeru (P3Vx)
    # pravoj kroz P8 u y-smeru (P8Vy)
    l1 = prava_kroz(P3, BeskonacnoX)
    l2 = prava_kroz(P8, BeskonacnoY)

    P4 = presek_pravih(l1,l2)

    return P4
```

```python
lista = [p1, p2, p3, p5, p6, p7, p8]

P4 = osmoteme(lista)
print(P4)
```
